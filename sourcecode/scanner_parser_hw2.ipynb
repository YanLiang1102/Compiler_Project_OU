```python
#the index will be the start of of the empty space.
def findNoOfSpaces(inputdata,index):
    count=0;
    cur=inputdata[index]
    while(cur==" "):
        count=count+1;
        index=index+1;
        cur=inputdata[index]
    return count;    
    
```

```python
symboltable={}
def bookkeeper(word,wordtype):
    if not word in symboltable.keys():
        symboltable[word]=wordtype;
```

```python
#so if the token is recognized can not be a keyword then need to skip everything behind!
def skipCharacters(input,index,token):
    #if the end id not seperator then keep going:
    cur=input[index]
    #token=token+cur;
    length=1;
    while (cur!=" " and cur!="," and cur!=";" and cur!=":" and cur!="!" and cur!="?" and cur!="(" and cur!=")" and cur!="\n") and index<len(input)-1:
        #print(index)
        token=token+cur
        index=index+1;
        cur=input[index]
        length=length+1
    #need to trim the last special character!
    return {"token":token,"length":int(length)-1}
    
```

```python
nexttoken=[];
def skipAll(keyword,cur,inputdata,i,token):
    if(cur==" " or cur=="," or cur==":" or cur=="!" or cur=="?" or cur==";" or cur=='(' or cur==')'or cur=='\n'):
                                        nexttoken.append([keyword,"keyword"])
                                        #print(keyword+ " keyword")
                                        return 0;
    else:
        result=skipCharacters(inputdata,i,token)
        #print(result["token"]+" identifier")
        nexttoken.append([result["token"],"identifier"])
        #also need to add into the symbol table
        bookkeeper(result["token"],"id")
        return result["length"]-1
```

```python
def skipNoMatterWhat(cur,inputdata,i):
    count=0;
    while(cur!=" "):
        count=count+1;
        i=i+1;
        cur=inputdata[i]
    return count;
```

```python
def skipUntilLineBreak(inputdata,i):
    count=0;
    cur=""
    while(cur!="\n"):
        count=count+1;
        i=i+1;
        cur=inputdata[i]
    return count;
```

```python
class Stack:
  def __init__(self):
    self.__storage = []

  def isEmpty(self):
    return len(self.__storage) == 0

  def push(self,p):
    self.__storage.append(p)

  def pop(self):
    return self.__storage.pop()
  def peek(self):
    return self.__storage[len(self.__storage)-1]
```

```python

def parse():
    print("step"+"   "+"top"+" "*13+"ahead"+" "*11+"action");
    st=0;
    scannercount=0;
    stack=Stack();
    fetchinput=True;
    stack.push("trump");
    print("# 0"+"   "+"----"+" "*13+"----"+" "*11+"push trump");
    #print("# 0          --          --          push trump");
    debug="";
    while(st<len(inputdata) and not stack.isEmpty()):
        top=stack.peek();
        scannercount=scannercount+1;
        if(fetchinput):
            st=scanner(inputdata,symboltable,st);
            while(len(nexttoken)==0):
                st=scanner(inputdata,symboltable,st);
            token=nexttoken[0][0];
            debug=token;
            #parse just need to know the type of identifier, constants and string.
            if(nexttoken[0][1]=="identifier"):
                token="identifier"
            elif(nexttoken[0][1]=="constants"):
                token="const"
            elif(nexttoken[0][1]=="string"):
                token="string"
        #rule=getRule(top,token);
        if(top in nonterminals):
            #need to check the rule
            #print("top and lookahead: "+str(top)+"  "+str(token))
            ruleindex=getRule(top,token);
            #the fetch the rule and push the cporrespond stuff on to stack
            print("# "+str(scannercount)+"  "+top+" "*(14-len(top))+token+" "*(14-len(token))+"  rule:"+ str(ruleindex));
            rule=rules[ruleindex]
            #the push the stuff from the end to the stack:
            size=len(rule)-1;
            #0 will not included since that is the non-termianl on the left handside,
            #we only need to push the right handside
            stack.pop();
            for n in range(size,0,-1):
                if(rule[n]=="epsilon"):
                    #fetchinput=False;
                    continue;
                stack.push(rule[n]);
            fetchinput=False; #since not consume yet
        else:
            #check match
            if(top==token):
                stack.pop()
                print("# "+str(scannercount)+"  "+top+" "*(14-len(top))+token+" "*(14-len(token))+"  match & pop "+str(debug));
                #print("# "+str(scannercount)+"  "+top+" "*(10-len(top))+token+"+" "*(10-len(token))+"  rule: "+str(ruleindex));
                #print("# "+str(scannercount)+"          "+top+"          "+token+"          match & pop "+str(debug));
                #input consumed need to fetch more
                fetchinput=True;
            else:
                #print("top: "+top)
                #print("lookahead: "+token)
                #print("debug "+debug);
                print("error! and stop!")
                while(not stack.isEmpty()):
                    print(stack.pop())
                return;
        #print("debug: "+debug)    
```
